```python
import numpy as np
import pandas as pd

from iterstrat.ml_stratifiers import MultilabelStratifiedKFold
```

```python
train_df = pd.read_csv("train_encoded.csv")
test_df  = pd.read_csv("test_encoded.csv")
```

```python
y = train_df.drop(['risk_flag'],axis=1)

kf = MultilabelStratifiedKFold(n_splits=9, shuffle=True , random_state=0)


for f, (t_, v_) in enumerate(kf.split(X=train_df, y=y)):
    train_df.loc[v_, 'kfold'] = f
```

```python
train_df.kfold = train_df.kfold.astype(np.int64)
```

```python
train_df[train_df.kfold ==9].risk_flag.sum()
```

```python
train_df.to_csv("train_folds.csv",index=False)
```
